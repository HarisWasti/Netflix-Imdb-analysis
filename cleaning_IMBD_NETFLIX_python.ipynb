## Cleaning up the dataset

### Remove the extra columns

```python
df = df.drop(['index', 'id', 'imdb_id'], axis = 1)
df
```

### What rows have at least one missing value?

```python
nan_df = df[df.isna().any(axis=1)]
nan_df
```

### How many missing values are their in the column 'Description'?

```python
description_nan_count = df['description'].isna().sum()
description_nan_count
```

### How many missing values are their in the column 'Imdb votes'?

```python
imdb_votes_nan_count = df['imdb_votes'].isna().sum()
imdb_votes_nan_count
```

### How many missing values are their in the column 'age certification'?

```python
age_certification_nan_count = df['age_certification'].isna().sum()
age_certification_nan_count
```

### What is the average No. of votes?

```python
median_imdb_votes = df['imdb_votes'].median()
median_imdb_votes
```

### Fill Na with valid placement

```python
df['description'].fillna('No description', inplace = True)
```

```python
df['imdb_votes'].fillna(median_imdb_votes, inplace = True)
```

```python
df['age_certification'].fillna('No certification', inplace = True)
```

## Univariate analysis

```python
df
```

### Highest rated and lowest rated

```python
highest_rated = df.sort_values('imdb_score', ascending=False)
highest_rated
```

### Highest rated and lowest rated but must have more than 1000 votes

```python
rating_filtered_df = df[df['imdb_votes'] > 1000].sort_values(by='imdb_score', ascending=False)
rating_filtered_df
```

### Highest and lowest rated movies 

```python
rating_movie_df = df[(df['imdb_votes'] > 1000) & (df['type'] == 'MOVIE')].sort_values(by='imdb_score', ascending=False)
rating_movie_df
```

### Statistics for movies

```python
rating_movie_df.describe()
```

### Highest and lowest rated shows

```python
rating_show_df = df[(df['imdb_votes'] > 1000) & (df['type'] == 'SHOW')].sort_values(by='imdb_score', ascending=False)
rating_show_df
```

### Statistics for shows

```python
rating_show_df.describe()
```

### shows/movies with 0 listed runtime

```python
run_time = df[df['runtime'] == 0]
run_time
```
